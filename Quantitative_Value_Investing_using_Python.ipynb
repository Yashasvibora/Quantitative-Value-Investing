```python
!pip install yfinance
!pip install talib-binary

import io
import pandas as pd
import yfinance as yf
import requests 
import talib

def getdata(url):
  r = requests.get(url,headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
  data = pd.read_html(r.text)
  return data

ticker= "GOOGL"

statistics_url = f'https://finance.yahoo.com/quote/{ticker}/key-statistics?p={ticker}'
statistics_data = getdata(statistics_url)

df = yf.download("GOOGL", start="2018-11-01", end="2022-10-18", interval="1d")
df.head()
```

```python
df.loc[:, 'ma20'] = df.Close.rolling(20).mean()
df.loc[:, 'ma200'] = df.Close.rolling(200).mean()
df[["Close", "ma20", "ma200"]].plot()
```

```python
import talib
import matplotlib.pyplot as plt

df.loc[:, "rsi"] = talib.RSI(df.Close, 14)

fig, ax = plt.subplots(1, 2, figsize=(30, 7))
ax0 = df[["rsi"]].plot(ax=ax[0])
ax0.axhline(30, color="black")
ax0.axhline(70, color="black")
df[["Close"]].plot(ax=ax[1])
```
